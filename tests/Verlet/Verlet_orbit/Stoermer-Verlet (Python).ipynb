# Algoritmo de Verlet

$X(t+dt)=2X(t)-X(T-dt)+A(t)dt^2$


```python
import numpy as np
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Distancia: UA
#Tiempo: días
#Aceleración: UA^2/Días

#Constante gravitacional
MS=1;
MJ=MS/1048;
G=39.485279383758936; # G=4*pi^2 Y^-2*AU*MS^-1
#G=6.67*10**-11*(6.68*10**-12)**3/(5*10**-31*(1.15*10**-5)**2);
#Condiciones iniciales

r1=[0.0348,-0.5315723,-0.2304572]; #Posición partícula 1 en t1
r2=[0,0,0]; #Posición partícula 2 en t1
v=[0.00956703,0.01288496,0.00558611]; #Velocidad partícula 1 en t1
#r1=[1,0]
#v=[0,0.01721420632]
dt=0.1; #Intervalo de tiempo
n=500000; #Número de iteraciones
```

```python
#Norma del vector posición

def norma(a,b):
    s=0;

    for i in range (0,len(a),1):
        s=s+(a[i]-b[i])**2;
    norm=np.sqrt(s);
    return norm**3;
```

```python
#Aceleraciones

def Acel(a,b):
    global G,MS;

    R=norma(a,b);
    A=[];

    for i in range (0,len(a),1):
        ac=-G*MS*(a[i]-b[i])/R;
        A.append(ac);
    return A;
```

```python
#Condiciones después de un intervalo de tiempo dt
#rp = posición posterior

def pospos(a,b):
    global dt;

    rp=[];

    for i in range (0,len(a),1):
        RP=a[i]+b[i]*dt;
        rp.append(RP);
    return rp;
```

```python
def Verlet():
    global r1,r2,dt,v;
    
    rp=pospos(r1,v);
    a=Acel(rp,r2);

    x1=r1[0];
    y1=r1[1];
    z1=r1[2];

    x2=rp[0];
    y2=rp[1];
    z2=rp[2];

    x=[x1,x2];
    y=[y1,y2];
    z=[z1,z2];

    for i in range (0,n,1):
        xn=2*rp[0]-r1[0]+a[0]*dt**2;
        yn=2*rp[1]-r1[1]+a[1]*dt**2;
        zn=2*rp[2]-r1[2]+a[2]*dt**2;
        x.append(xn);
        y.append(yn);
        z.append(zn);
        r1=rp;
        rp=[xn,yn,zn];
        a=Acel(rp,r2);
        

    return x,y,z;
    #return x,y;
```

```python
R = Verlet()
```
