```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
# Reading the data from a text file
data1 = open ("coordinates1.txt","r");
x1_pts = data1.readlines();

data2 = open ("coordinates2.txt","r");
x2_pts = data2.readlines();

# Creating the x and y arrays
x1 =[];
y1 =[];
z1 =[];

x2 =[];
y2 =[];
z2 =[];

for line in x1_pts:
    
    if (line.split()[0]!='x' or line.split()[1]!='y' or line.split()[2]!='z'):
        x1.append(float(line.split()[0]));
        y1.append(float(line.split()[1]));
        z1.append(float(line.split()[2]));
for line in x2_pts:
    
    if (line.split()[0]!='x' or line.split()[1]!='y' or line.split()[2]!='z'):
        x2.append(float(line.split()[0]));
        y2.append(float(line.split()[1]));
        z2.append(float(line.split()[2]));
```

```python
len (x1)
```

```python
fig = plt.figure();
ax = fig.add_subplot(projection='3d');

for i in range (0,len(x1),1):
    
    ax.scatter(x1[i], y1[i], z1[i], marker="*", color = "rebeccapurple");
    ax.scatter(x2[i], y2[i], z2[i], marker="o", color = "gold", label = "Sun");
    
ax.set_xlabel('x');
ax.set_ylabel('y');
ax.set_zlabel('z');
plt.show()
```
