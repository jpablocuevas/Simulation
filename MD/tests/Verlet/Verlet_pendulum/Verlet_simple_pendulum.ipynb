```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
# Reading the data from a text file
data = open ("position.txt","r");
pts= data.readlines();

# Creating the x and y arrays
x =[];
y =[];

for line in pts:
    
    if (line.split()[0]!='x' or line.split()[1]!='y'):
        x.append(float(line.split()[0]));
        y.append(float(line.split()[1]));
```

```python
len (x)
```

```python
fig, ax = plt.subplots()
ax.scatter(x, y, c="darkblue", label = "Simple pendulum",alpha=0.3, edgecolors='none', marker = "o")
ax.legend()
ax.grid(True)
plt.show()
```
